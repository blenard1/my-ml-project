# Salient Object Detection Demo

Run the cells after training a checkpoint with `train.py`. The notebook loads the model, accepts an image path, and shows the input, predicted mask, overlay, and inference time.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import torch
from PIL import Image

from evaluate import load_model_from_checkpoint, predict_image

checkpoint_path = Path('checkpoints/best/best_model.pt')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model, config = load_model_from_checkpoint(checkpoint_path, device)
print(config)
```

```python
image_path = Path('sample.jpg')  # replace with your test image
image = Image.open(image_path).convert('RGB')
mask, overlay, elapsed_ms = predict_image(model, image, device, image_size=config.get('image_size', 128), threshold=0.40)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(image.resize((config.get('image_size', 128), config.get('image_size', 128))))
axes[0].set_title('Input image')
axes[1].imshow(mask, cmap='gray', vmin=0, vmax=1)
axes[1].set_title('Predicted saliency mask')
axes[2].imshow(overlay)
axes[2].set_title(f'Overlay ({elapsed_ms:.1f} ms)')
for ax in axes:
    ax.axis('off')
plt.show()
```
